# Loc and ILoc Subsetting

```python
import pandas as pd

df = pd.read_csv('../00_datasets/titanic.csv')
df
```

```python
# use of loc function
# df[df['sex','age']]
df.loc[0:5, 'sex':'fare']

```

```python
# fare to deck
# df[df['fare', 'embarked', 'class', 'who', 'adult_male', 'deck']]
df.loc[0:15, 'fare':'deck']
```

```python
# negative index can be used
df.loc[-50:10, 'fare':'deck']
```

```python
# using iloc function
df.iloc[0:5, 0:3]
```
